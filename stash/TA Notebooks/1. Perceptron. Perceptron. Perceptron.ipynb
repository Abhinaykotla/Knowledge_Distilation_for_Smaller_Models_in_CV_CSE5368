```python
import pandas as pd
from sklearn.utils import shuffle
import numpy as np
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', names=['feat_0', 'feat_1', 'feat_2', 'feat_3', 'label'])
# df.columns = ['feat_0', 'feat_1', 'feat_2', 'feat_3', 'label']
```

```python
df
```

```python
features = df[['feat_0', 'feat_1', 'feat_2', 'feat_3']].values
labels = df['label'].values
```

```python
df['label'].value_counts()
```

```python
def get_prediction(x, w):
  a = np.dot(x, w)
  return 1 if a > 0.0 else -1


np.random.seed(1)
n_epochs = 15
w = np.random.random(5)
eta = 0.001
n_data = len(labels)

for epoch_idx in range(n_epochs):
  n_accurate = 0
  for f, label in zip(features, labels):
    x = np.hstack([1, f])
    y = 1 if label == 'Iris-setosa' else -1
    y_pred = get_prediction(x, w)
    if y == y_pred:
      n_accurate += 1
    w = w + eta*(y-y_pred)*x
  print(f'After epoch {epoch_idx}: Accuracy {100*n_accurate/n_data:0.2f}% with weights {[round(w_i, 3) for w_i in w.tolist()]}')


```
